```python
####################
#set up packages 
####################
import ee                     
ee.Initialize()
import pandas as pd
import geemap.eefolium as emap
import csv
import time
from datetime import date
import numpy as np
import os, sys
import math
```

```python
#map set up
Map=emap.Map()
Map.setOptions('SATELLITE')
countries = ee.FeatureCollection('FAO/GAUL/2015/level0') #countries is the dataset
#add countries
countryselected = ['Cuba','Antigua and Barbuda','Bahamas','Barbados','Belize','Aruba','Dominican Republic','Haiti','Puerto Rico','Cayman Islands','Dominica','Jamaica','Trinidad and Tobago','Saint Vincent and the Grenadines','Saint Kitts and Nevis','Saint Lucia']#'Turks and Caicos', 
countryCollection = countries.filter(ee.Filter.inList('ADM0_NAME', countryselected))
Map.addLayer(countryCollection,{'color': 'purple'},'country')

```

```python
files=os.listdir('Hurricane_Tracks')
for file in files:

    #select storm
    data0 = pd.read_csv("Hurricane_Tracks/"+file)
    data0
    #find wind
    maxwind=data0.WMO_WIND.max()
    
    data = data0[['LON','LAT']]
    coors = data.values.tolist()
    #sort wind
    if maxwind >=157:
        colors = 'blue'
        

    if maxwind >=130 and maxwind <157:
        colors = 'red'

    if maxwind <111:
        colors = 'green'
    
    #add wind
    track = ee.Geometry.LineString(coors)

    Map.addLayer(track,{'color': colors},file)
    
```
